```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Plot settings
plt.rc("font", family=["Helvetica", "Arial"]) # skifter skrifttype
plt.rc("axes", labelsize=16)   # skriftstørrelse af `xlabel` og `ylabel`
plt.rc("xtick", labelsize=16, top=True, direction="in")  # skriftstørrelse af ticks og viser ticks øverst
plt.rc("ytick", labelsize=16, right=True, direction="in")
plt.rc("axes", titlesize=16)
plt.rc("legend", fontsize=16)
```

```python
# Load streeteasy data-set
df = pd.read_csv("https://raw.githubusercontent.com/sonnynomnom/Codecademy-Machine-Learning-Fundamentals/master/StreetEasy/manhattan.csv")

# Define features and label
x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs',
        'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator',
        'has_dishwasher', 'has_patio', 'has_gym']]
y = df[['rent']]

# Split data in training and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8,  
                                                    test_size = 0.2, random_state=6)

# Create model and fit to data
lm = LinearRegression()
model = lm.fit(x_train, y_train)

# Make predictions
y_predict= lm.predict(x_test)

# Print training and test scores
print("Train score:")
print(lm.score(x_train, y_train))
print("Test score:")
print(lm.score(x_test, y_test))

# Make scatter-plot of test vs. predictions
plt.scatter(y_test, y_predict, alpha=0.4)
plt.plot(range(20000), range(20000))

# Label plot and show
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Actual Rent vs Predicted Rent")
plt.show()
```
